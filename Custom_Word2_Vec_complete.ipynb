## Custom Training for Word2Vec

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import gensim
import nltk
```

```python
nltk.download("all")
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []
```

```python
for file in os.listdir("/content/drive/MyDrive/GEN AI ineuron/got_data"):
  print(file)
```

```python
f = open(r"/content/drive/MyDrive/GEN AI ineuron/got_data/004ssb.txt",encoding = "unicode_escape")
corpus = f.read()
raw_sent = sent_tokenize(corpus)
```

```python
for i in raw_sent:
  print(simple_preprocess(i))
```

```python
for filename in os.listdir("/content/drive/MyDrive/GEN AI ineuron/got_data"):
  file_path = os.path.join("/content/drive/MyDrive/GEN AI ineuron/got_data",filename)
  with open(file_path,encoding  = "unicode_escape") as f:
    corpus = f.read()
  raw_sent = sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
model  = gensim.models.Word2Vec(window = 5,min_count = 2,vector_size = 4)
```

```python
model.build_vocab(story)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.corpus_total_words
```

```python
model.min_count
```

```python
model.train(story, total_examples = model.corpus_count, epochs = model.epochs)
```
